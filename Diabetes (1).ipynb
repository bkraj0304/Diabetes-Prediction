# Step Outlier Treatments

```python
# we dont do outlier treatment for target as well as categorical variable
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
wisker(df['BloodPressure'])
```

```python
df.columns
```

```python
for i in['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness','BMI','Age']:
        lw,uw=wisker(df[i])
        df[i]=np.where(df[i]<lw,lw,df[i])
        df[i]=np.where(df[i]>uw,uw,df[i])
```

# Handling  multi-collinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']

# Calculate VIF
X = X[features]
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
#choosing our dependent and independent variables,respectively
y = df['Outcome']
x = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Choosing training data to be 30% and rest as our testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#creating object for Logistic regression
lr_under = LogisticRegression()
```

```python
#providing datas to train the machine
lr_under.fit(x_train,y_train)
```

```python
#storing the predicted values
y_under_pred = lr_under.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#printing the classification report
print(classification_report(y_test,y_under_pred))
```

```python
#printing the confusion matrix
print(confusion_matrix(y_test,y_under_pred))
```
